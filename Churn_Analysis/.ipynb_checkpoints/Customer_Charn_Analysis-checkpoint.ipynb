# converted 0 and 1 value of senior citizens to yes/no to make it easier to understand 

```python
def convert(value):
    if value == 1:
        return "yes"
    else :
        return "no"

df["SeniorCitizen"]=df["SeniorCitizen"].apply(convert)
df["SeniorCitizen"][20]
```

```python
ax=sns.countplot(x="Churn",data=df)
ax.bar_label(ax.containers[0])
plt.show()
print(ax.containers[0])
```

# percentage of churned Customers

```python
plt.figure(figsize=(3,4))
gb=df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb["Churn"],labels=gb.index,autopct="%1.3f%%")
plt.title("Percentage of Churned Customers",fontsize=10)
plt.show()
```

# from the given pie chart we can conclude that 26.54% of our customers have churned out. not let's explore the reason behind it

```python
plt.figure(figsize=(4,4))
ax=sns.countplot(x='gender',data=df,hue="Churn")
ax.bar_label(ax.containers[0])
plt.title("Churned by Gender")
plt.show()
```

```python
plt.figure(figsize=(4, 4))
ax = sns.countplot(x='SeniorCitizen', data=df, hue="Churn")
ax.bar_label(ax.containers[0])
plt.title("Churned by Senior Citizen Status")
plt.show()
```

```python
counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)
percentages = counts.div(counts.sum(axis=1), axis=0) * 100

# Create the plot
fig, ax = plt.subplots(figsize=(4, 4))
bottoms = [0] * len(percentages)

for churn_status in percentages.columns:
    ax.bar(percentages.index, percentages[churn_status], label=f"Churn: {churn_status}", bottom=bottoms)
    bottoms += percentages[churn_status]

# Add percentage labels
for idx, row in percentages.iterrows():
    cumulative = 0
    for churn_status in percentages.columns:
        percentage = row[churn_status]
        if percentage > 0:
            ax.text(idx, cumulative + percentage / 2, f"{percentage:.1f}%", ha="center", va="center")
        cumulative += percentage

# Formatting
ax.set_title("Churn Percentage by Senior Citizen Status")
ax.set_xlabel("Senior Citizen")
ax.set_ylabel("Percentage")
ax.legend(title="Churn Status")
plt.xticks(ticks=percentages.index, labels=["No", "Yes"])  # Assuming 0 = No, 1 = Yes
plt.show()
```

```python
# List of columns for count plots
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up the subplot grid
num_cols = 3  # Number of columns per row in the grid
num_rows = -(-len(columns) // num_cols)  # Calculate the number of rows (ceiling division)

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, num_rows * 4))
axes = axes.flatten()  # Flatten axes array for easy indexing

# Plot each count plot
for i, col in enumerate(columns):
    ax=sns.countplot(data=df, x=col, hue="Churn", ax=axes[i])
    ax.bar_label(ax.containers[0])
    axes[i].set_title(f"Churn by {col}")
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")

# Hide any unused subplots
for j in range(len(columns), len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```
