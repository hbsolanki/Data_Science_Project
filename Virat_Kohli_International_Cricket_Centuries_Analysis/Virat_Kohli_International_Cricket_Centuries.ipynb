```python
#import Lib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("./Dataset/Virat_Kohli_International_Cricket_Centuries.csv")
df
```

```python
#1 Bar chart (team vs run of cen)
team_and_runs={}
for i in df[["Against","Runs"]].values:
    i[1]=i[1].replace("*","")
    if i[0] in team_and_runs :
        team_and_runs[i[0]]+=int(i[1])
    else:
        team_and_runs[i[0]]=int(i[1])
       
plt.bar(list(team_and_runs.keys()),list(team_and_runs.values()))
plt.xlabel("team")
plt.ylabel("Runs of century")
plt.title("Team Vs runs by Century")

# Adjust spacing and rotation of x-axis labels
plt.xticks(ticks=range(len(team_and_runs.keys())), labels=team_and_runs.keys(), rotation=45, ha="right", fontsize=10)

# Show the plot
plt.tight_layout() 
plt.show()
```

```python
#2 Position vs Century

position_century_count={}

for i in df["Position"]:
    if i in position_century_count:
        position_century_count[i]+=1
    else : 
        position_century_count[i]=1

```
